```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style Configuration
sns.set_theme(style="whitegrid")
netflix_red = "#E50914" # Corporative color of netflix

# 1. Load data
df = pd.read_csv('../data/netflix_data.csv')

# --- Chart 1: Pelculas vs Series ---
plt.figure(figsize=(7, 7))
df['type'].value_counts().plot(
    kind='pie', 
    autopct='%1.1f%%', 
    colors=[netflix_red, '#564d4d'], 
    startangle=90,
    explode=(0.05, 0)
)
plt.title('Distribution of Netflix Content', fontsize=15, pad=20)
plt.ylabel('')
plt.savefig('../img/content_type_distribution.png', bbox_inches='tight')
plt.show()

# --- Chart 2: Top 10 Countries---
plt.figure(figsize=(10, 6))
# Clean some of the data of the countries (take the first if there is various)
top_countries = df['country'].str.split(', ').str[0].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index, palette='Reds_r')
plt.title('Top 10 Countries Productors of Content', fontsize=15)
plt.xlabel('Number of Titles')
plt.ylabel('Countries')
plt.savefig('../img/top_10_countries_bar.png', bbox_inches='tight')
plt.show()
```
